# Feature Engineering

```python
# Importing pandas
import pandas as pd
import numpy
```

```python
# reading file
df =pd.read_excel("Before Preprocessing.xlsx")
df
```

```python
# Missing Values
class Imp:
    def __init__(self,dataframe ,strategy ="mean"):
        self.dataframe =dataframe
        self.strategy =strategy
        
    
    def fitandTransfrom(self):
        for i in self.dataframe.columns:
            if type(self.dataframe.loc[0, i])==str:
                self.dataframe.loc[self.dataframe[i].isnull()==True,  i] =self.dataframe[i].mode()[0]
            elif self.strategy=="mean":
                
            
                self.dataframe.loc[self.dataframe[i].isnull()==True,  i] =self.dataframe[i].mean()
            else:
                self.dataframe.loc[self.dataframe[i].isnull()==True,  i] =self.dataframe[i].median()
            
        return df
      
class LabelEncoding:
    def __init__(self, dataframe, skip_last =True):
        self.skip_last =skip_last
        self.dataframe =dataframe
        self.size =len(self.dataframe.columns)

    def fitandTransfrom(self):
        k =0
        if self.skip_last==True:
            k =1
        l=[]
        for i in range(self.size-k):
            v =self.dataframe.columns[i]
            if type(self.dataframe.loc[0, v])==str:
                for k in self.dataframe[v].unique():
                    self.dataframe.loc[:, k] =0
                    self.dataframe.loc[self.dataframe[v]==k, k] =1
                l.append(v)

        for i in l:
            self.dataframe =self.dataframe.drop(i, axis =1)

        return self.dataframe


class FeatureScaling:
    def __init__(self,dataframe, skip_first):
        self.dataframe =dataframe
        self.skip_first =skip_first
    def fitandTransfrom(self):
        if self.skip_first==True:
            k =1
        else:
            k =0
        for i in range(k, len(self.dataframe.columns)):
            v =self.dataframe.columns[i]
            if type(self.dataframe.loc[0, v]) in [numpy.float64, numpy.int64]:
                maxi =self.dataframe[v].max()
                mini =self.dataframe[v].min()
                self.dataframe.loc[:, v] =(self.dataframe.loc[:,v]-mini)/(maxi -mini)

        return self.dataframe
```

```python
# Handling Missing Values
imp =Imp(df, "mean")
df =imp.fitandTransfrom()
df





```
