## Reading CSV

## Reading TSV

```python
df =pd.read_csv("movie_titles_metadata.tsv", sep="\t", names=["sno", "name", "releaseyear", "rating", "votes", "genres"])
df
```

## Index_col

```python
# To make some col as index one
df =pd.read_csv("score.csv", index_col="Hours")
df
```

## Header(to make specific row header)

```python
df =pd.read_csv("test.csv", header=0)
df
```

## Use_cols (to select specific cols)

```python
df =pd.read_csv("test.csv", usecols=["enrollee_id", "education_level", "company_type"])
df
```

## Skip rows

```python
# Will skip nth rows while reading data from file
df =pd.read_csv("test.csv", skiprows=[1,3])
df
```

## Changing dtype of cols

```python
df =pd.read_csv("aug_train.csv", dtype={"target": "int8"})
df.info()
```

## Handling Dates

```python
df =pd.read_csv("IPL Matches 2008-2020.csv")
df.info()
```

```python
df =pd.read_csv("IPL Matches 2008-2020.csv", parse_dates=["date"])
df.info()
```

## Converters

```python
def rename(name):
    if name=="Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python

df =pd.read_csv("IPL Matches 2008-2020.csv", converters={"team1": rename})
df
```

## Na_values (to make some specific val in col nan)

```python
df =pd.read_csv("IPL Matches 2008-2020.csv", na_values=["Bangalore"])
df
```
